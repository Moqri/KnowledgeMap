```python
import pandas as pd
import glob
```

```python
in_path =r'../../data/raw/' # use your path
out_path= "../../data/clean/"
allFiles = glob.glob(in_path + "/*.txt")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_table(file_,index_col=None, header=0)
    list_.append(df)
frame = pd.concat(list_)
frame.to_csv(out_path+"concat.csv")
```
